```python
from collections import Counter
import matplotlib.pyplot as plt 
counts = Counter(cyphertext)
total = sum(counts.values())

# Sortieren nach Häufigkeit
sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)

# Prozent berechnen
letters = []
percentages = []

for letter, count in sorted_counts:
    percent = (count / total) * 100
    letters.append(letter)
    percentages.append(percent)



# Plot
plt.figure()
plt.bar(letters, percentages)
plt.xlabel("Buchstaben")
plt.ylabel("Prozent (%)")
plt.title("Buchstabenhäufigkeit in Prozent")
plt.show()

```

```python
counts = Counter("""IN CRYPTOGRAPHY, A CAESAR CIPHER IS ONE OF THE SIMPLEST AND MOST
WIDELY KNOWN ENCRYPTION TECHNIQUES. IT IS A TYPE OF SUBSTITUTION
CIPHER IN WHICH EACH LETTER IN THE PLAINTEXT IS REPLACED BY A
LETTER SOME FIXED NUMBER OF POSITIONS DOWN THE ALPHABET. IT IS
UNKNOWN HOW EFFECTIVE THE CAESAR CIPHER WAS AT THE TIME, BUT IT
IS LIKELY TO HAVE BEEN REASONABLY SECURE, NOT LEAST BECAUSE
MOST OF CAESAR’S ENEMIES WOULD HAVE BEEN ILLITERATE AND OTHERS
WOULD HAVE ASSUMED THAT THE MESSAGES WERE WRITTEN IN AN UNKN-
OWN FOREIGN LANGUAGE. EVEN AS LATE AS 1915, THE CAESAR CIPHER WAS
IN USE: THE RUSSIAN ARMY EMPLOYED IT AS A REPLACEMENT FOR MORE
COMPLICATED CIPHERS WHICH HAD PROVED TO BE TOO DIFFICULT FOR THEIR
TROOPS TO MASTER.""")

total = sum(counts.values())

# Wahrscheinlichkeiten berechnen
probabilities = {char: count / total for char, count in counts.items()}

# sortiert nach Wahrscheinlichkeit
sorted_probs = sorted(probabilities.items(), key=lambda x: x[1], reverse=True)

for char, prob in sorted_probs:
    print(f"{char}: {prob:.4f}")
```
