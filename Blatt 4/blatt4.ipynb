```python
def f1(x, k):
    return SBOX[(x + k) % 256]

def f2(x, k):
    return INV_SBOX[(x + k) % 256]
```

```python
def encrypt(m0, m1, k): 
    l2 = f1(m1,k)^m0 
    r2 = f2(f1(m1,k)^m0,k)^m1 
    return l2, r2 

def decrypt(l2, r2, k): 
    m1 = f2(l2, k) ^r2 
    m0 = f1(f2(l2,k)^r2,k)^l2
    return m0, m1
```

```python
import random

def count_bits(x):
    return bin(x).count("1")

def confusion_test(num_tests=100):
    difs = []

    for _ in range(num_tests):
        m0 = random.randint(0, 255)
        m1 = random.randint(0, 255)

        key_base = random.randint(0, 255)

        bit_to_flip = 1 << random.randint(0, 7)
        key_prime = key_base ^ bit_to_flip

        
        c_base = encrypt(m0, m1, key_base)
        c_prime = encrypt(m0, m1, key_prime)

        diff = count_bits(c_base[0] ^ c_prime[0]) + \
               count_bits(c_base[1] ^ c_prime[1])

        difs.append(diff)

    # Durchschnitt berechnen
    avg = sum(difs) / len(difs)
    return avg, difs
```

```python
import matplotlib.pyplot as plt

avg, difs = confusion_test(200)

print("Average bit difference:", avg)

# --------- HISTOGRAMM ----------
plt.figure(figsize=(10,5))

plt.hist(difs, bins=range(min(difs), max(difs) + 2),
         edgecolor='black', alpha=0.75)

plt.axvline(avg, color='red', linestyle='--', label=f"Avg = {avg:.2f}")

plt.title("Confusion Test - Distribution of Bit Differences")
plt.xlabel("Bit differences")
plt.ylabel("Number tests")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def count_bits(x):
    return bin(x).count("1")

def diffusion_test(num_tests=100):
    difs = []

    for _ in range(num_tests):
        # gleiche Message
        m0 = random.randint(0, 255)
        m1 = random.randint(0, 255)

        # 1 Bit in der Message flippen
        bit_to_flip = 1 << random.randint(0, 7)

        m0_prime = m0
        m1_prime = m1

        # zufällig entscheiden, welches Byte geändert wird
        if random.random() < 0.5:
            m0_prime ^= bit_to_flip
        else:
            m1_prime ^= bit_to_flip

        # Schlüssel bleibt gleich (!! wichtig für diffusion)
        key = random.randint(0, 255)

        c_base = encrypt(m0, m1, key)
        c_prime = encrypt(m0_prime, m1_prime, key)

        diff = count_bits(c_base[0] ^ c_prime[0]) + \
               count_bits(c_base[1] ^ c_prime[1])

        difs.append(diff)

    avg = sum(difs) / len(difs)
    return avg, difs


# --------- AUSFÜHREN ----------
avg, difs = diffusion_test(200)

print("Average bit difference (diffusion):", avg)

# --------- HISTOGRAMM ----------
plt.figure(figsize=(10,5))

plt.hist(difs, bins=range(min(difs), max(difs) + 2),
         edgecolor='black', alpha=0.75)

plt.axvline(avg, color='red', linestyle='--', label=f"Avg = {avg:.2f}")

plt.title("Diffusion Test - 1 Bit Message Change")
plt.xlabel("Bit differences")
plt.ylabel("Number of tests")
plt.legend()
plt.grid(True)

plt.show()
```
